```python
from collections import deque


def solution(numbers, direction):
    numbers = deque(numbers)
    if direction == 'right':
        numbers.rotate(1)
    else:
        numbers.rotate(-1)
    return list(numbers)
```

```python
# https://school.programmers.co.kr/learn/courses/30/lessons/120956
import re

def solution(babbling):
    count = 0
    p = ["aya", "ye", "woo", "ma"]
    for i in babbling:
        print('-------------')
        print(i)
        for pattern in p:
            i = re.sub(pattern, ' ', i)
            print('--p--')
            print(i)
        if i.replace(' ', '') == '':
            count += 1
    return count


solution(['mayaa', "ayaye", "uuuma", "ye", "yemawoo", "ayaa"])


# 'mayaa' => 'm a' => 카운트가 안됨
```
